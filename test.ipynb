```python
import pandas as pd

# Baca file excel
df = pd.read_excel("C:\\Users\\abednego\\Downloads\\Form-Sales-Force-Effectiveness\\List Produck listing.xlsx"
, sheet_name="Sheet1")

# Pastikan kolom sesuai (ubah kalau header beda di file excel)
# Misalnya header: Procode, Itname, HNA, CostCenter
df = df.rename(columns={
    "Procode": "procode",
    "Itname": "itname",
    "HNA": "hna",
    "CostCenter": "costCenter"
})

# Buat array JavaScript
produkData = []
for _, row in df.iterrows():
    produkData.append({
        "procode": str(row["procode"]).zfill(7),  # pastikan 7 digit
        "itname": row["itname"],
        "hna": int(row["hna"]),
        "costCenter": row["costCenter"]
    })

# Cetak dalam format JavaScript
print("const produkData = [")
for item in produkData:
    print("  {")
    print(f'    procode: "{item["procode"]}",')
    print(f'    itname: "{item["itname"]}",')
    print(f'    hna: {item["hna"]},')
    print(f'    costCenter: "{item["costCenter"]}",')
    print("  },")
print("];")

# simpan jadi file .js
with open("produkData.js", "w", encoding="utf-8") as f:
    f.write("const produkData = ")
    f.write(str(produkData).replace("'", '"'))  # biar format JSON
    f.write(";")
```
